```python
#Creating truncating function to eliminate outliers
#Observations beyond of the interquartile range are imputed or replaced with the maximum and minimum value of the distibution
def truncate(Ys):
    
    Q1 = Ys.quantile(0.25)
    Q3 = Ys.quantile(0.75)
    IQR = Q3 - Q1
    upper_bound =  Q3 + (1.5 * IQR)
    lower_bound = Q1 - (1.5 * IQR)
    value_outlier_Y_up = Ys > upper_bound
    value_outlier_Y_low = Ys < lower_bound
    outlier_listUB=len(Ys[value_outlier_Y_up])
    outlier_listLB=len(Ys[value_outlier_Y_low])
    if outlier_listUB != 0:
        Ynew = Ys.replace([Ys[value_outlier_Y_up]], upper_bound)
    if outlier_listLB != 0:
        Ynew = Ynew.replace([Ynew[value_outlier_Y_low]], lower_bound) 
    
    return Ynew
    
    
```

```python
#Applying trucating function on the data
data_clean_nn['age_days']=truncate(data_clean_nn['age_days'])
data_clean_nn['birds_present']=truncate(data_clean_nn['birds_present'])
data_clean_nn['mortality_birds']=truncate(data_clean_nn['mortality_birds'])
data_clean_nn['feed_intake_birds']=truncate(data_clean_nn['feed_intake_birds'])
data_clean_nn['feed_intake_per_bird_present']=truncate(data_clean_nn['feed_intake_per_bird_present'])
data_clean_nn['body_weight_bird']=truncate(data_clean_nn['body_weight_bird'])
```

```python
import seaborn as sns
```

```python
#Random split of 2/3 of the data was used as train set and 1/3 for the test data
data_clean_s=data_clean_nn.sample(frac=1).reset_index(drop=True)
train=data_clean_s[0:21789]
test=data_clean_s[21790:len(data_clean_s)]
```

```python
print("Size of train set : ",len(train))
print("Size of test set : ", len(test))
```

```python
#Libraries to perform linear mixed model
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
#Model formulas to be used in models ahead
fixed_formula="body_weight_bird ~ age_days + birds_present + mortality_birds + feed_intake_birds+feed_intake_per_bird_present"
rand_formula="~ age_days + birds_present + mortality_birds + feed_intake_birds+feed_intake_per_bird_present"
```

```python
#Linear mixed model with fixed intercept,fixed slope and random intercept
test_model_int = smf.mixedlm(fixed_formula, train, groups=train["flock_ID"].astype('category'))
test_model_fit_int=test_model_int.fit(reml=True)
```

```python
print(test_model_fit_int.summary())
```

```python
#Fitting mixed model with fixed slope and intercept + random slope and intercept
test_model_full = smf.mixedlm(fixed_formula, train, groups=train["flock_ID"].astype('category'),re_formula=rand_formula)
test_model_fit_full=test_model_full.fit(reml=True)
```

```python
print(test_model_fit_full.summary())
```

```python
#Extracting the random effects of model with random intercept
rand_effects=p.DataFrame.from_dict(test_model_fit_int.random_effects,orient='index')
#Extracting the random effects of model with random intercept and slope
rand_effects_full=p.DataFrame.from_dict(test_model_fit_full.random_effects,orient='index')
list(rand_effects)
rand_effects_full[1:5]
```

```python
#Library to plot QQplot
import pylab
```

# Model Assumptions

```python
#QQplot of the error terms and Random effects of model with random intercept
Res_int=test_model_fit_int.resid.values
Res_rand_int=rand_effects['groups']
print("Error terms")
sm.qqplot(Res_int, loc = 4, scale = 3, line='s')
pylab.show()
print("Random effects of intercept")
sm.qqplot(Res_rand_int, loc = 4, scale = 3, line='s')
pylab.show()
```

```python
#Extracting the error terms from model with random intercept and slope
Res_full=test_model_fit_full.resid.values

#QQplot of error terms
sm.qqplot(Res_full, loc = 4, scale = 3, line='s')
print("Error terms")
pylab.show()

sm.qqplot(rand_effects_full.Intercept, loc = 4, scale = 3, line='s')
print("Random intercept")
pylab.show()

sm.qqplot(rand_effects_full.age_days, loc = 4, scale = 3, line='s')
print("Random slope for age of chickens")
pylab.show()

sm.qqplot(rand_effects_full.feed_intake_per_bird_present, loc = 4, scale = 3, line='s')
print("Random slope for feed intake per chicken")
pylab.show()
```

```python
#In the figure obtained, move mouse pointer to highest bar to see the flock ID and its corresponding random effect
bar_data = [Bar(
            x=train.flock_ID.unique(),
            y=rand_effects.groups
    )]
layout=Layout(yaxis=dict(title='Random effects'),xaxis=dict(title='Flock ID'))
fig=Figure(data=bar_data,layout=layout)
plot(fig)
```
